```python
import pandas as pd
from pandasql import sqldf
```

```python
raw_data = pd.read_csv('programming_language.csv')
```

```python
raw_data.head()
```

```python
val_cols = raw_data.columns[1:]
```

```python
data = pd.melt(raw_data, id_vars='Date', value_vars = val_cols)
```

```python
data.head()
```

```python
data['Date'] = pd.to_datetime(data['Date'], format='%B %Y')
```

```python
data
```

```python
q = """

select 

strftime('%Y', Date) as year
, variable
, avg(value) as value


from Data

group by 1,2

;"""

df = sqldf(q, {'Data': data})
df
```

```python
df.to_csv('programming_languages_processed.csv', index=False)
```
